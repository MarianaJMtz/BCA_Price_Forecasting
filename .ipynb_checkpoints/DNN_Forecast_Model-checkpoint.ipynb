```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('BCA_MDA_ZONALPRICES.csv')
df.head()
```

```python
df_DNN = df.loc[:,['ZONE','HOUR','MDA LMP']]
df_DNN.head()
```

```python
df_DNN['DAY'] = df['DATE'].str[8:].apply(int)
df_DNN['MONTH'] = df['DATE'].str[5:7].apply(int)
```

```python
df_DNN.head()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df_DNN['ZONE'] = label_encoder.fit_transform(df_DNN['ZONE'])
```

```python
df_DNN.head(10)
```

```python
features = np.array(df_DNN.loc[:,['ZONE','HOUR','DAY','MONTH']])
prices = np.array(df_DNN.loc[:,'MDA LMP'])
prices = prices.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split

feat_train, feat_test, price_train, price_test = train_test_split(features, prices, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler

price_scaler = StandardScaler().fit(price_train)
```

```python
price_train_scaled = price_scaler.transform(price_train)
price_test_scaled = price_scaler.transform(price_test)
```

```python
feat_scaler = StandardScaler().fit(feat_train)
feat_train_scaled = price_scaler.transform(feat_train)
feat_test_scaled = price_scaler.transform(feat_test)
```

```python
from keras.utils import to_categorical

# One-hot encoding
feat_train_categorical = to_categorical(feat_train_scaled)
feat_test_categorical = to_categorical(feat_test_scaled)
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(units=8, kernel_initializer='normal', activation='relu', input_dim=4))
model.add(Dense(units=8, kernel_initializer='normal', activation='relu'))
model.add(Dense(units=1, kernel_initializer='normal', activation='linear'))
```

```python
# Compile the model
model.compile(optimizer='adam',
              loss='mean_squared_error',
              metrics=['accuracy'])
```

```python
feat_train_categorical = np.resize(feat_train_categorical, (85740, 4))
```

```python
model.fit(
    feat_train,
    price_train_scaled,
    epochs=3000,
    shuffle=True,
    verbose=2
)
```

```python
new_data = np.array([[0.9,0.0, 1.0, 0.2]])
print(f"Predicted class: {model.predict_classes(new_data)}")
```
